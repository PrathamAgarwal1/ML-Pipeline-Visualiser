# Exploratory Data Analysis

This notebook contains the Exploratory Data Analysis (EDA) functions used in the pipeline. It is imported directly by `app.py`.

```python
import numpy as np
import pandas as pd
import io
import base64
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Descriptive Statistics

```python
def generate_descriptive_stats(df):
    return df.describe().to_html(classes='table table-sm table-bordered')
```

### 2. Correlation Heatmap

```python
def generate_correlation_heatmap(df):
    corr = df.corr(numeric_only=True)
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    img = io.BytesIO()
    plt.savefig(img, format='png', bbox_inches='tight')
    img.seek(0)
    heatmap_url = base64.b64encode(img.getvalue()).decode()
    plt.close()
    return heatmap_url
```
